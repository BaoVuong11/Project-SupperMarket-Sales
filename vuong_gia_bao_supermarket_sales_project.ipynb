###Supermarket Sales Data Analysis


This dataset contains sales transactions from a supermarket across multiple branches in three major cities: Michigan, Chicago, and San Diego. The data includes key information about the sales transactions, customer demographics, and product details. Below is an overview of the columns present in the dataset:
Invoice ID:Unique identifier for each product.

Branch: Classification of supermarket branches.

City: Sales location.

Customer type: Classify customers as members of normal.

Gender: Male or female.

Product line: Name of each product.

Unit price: Price of the product.

Quantity: Number of products sold.

Tax: Tax index for each product.

Total revenue: Revenue generated from sales.

Date: Date of the sales record.

Time:Date of the sales record.

Payment: customer payment method.

Cogs: Sales amount before tax deduction.

Gross margin percentage: The index represents the profit for each product.

Gross income: profit for each product.

Rating: The index represents customer interest in each product line.
